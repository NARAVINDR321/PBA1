Probability distribution is a function that denotes the probability of all the outcomes of a random event.

A random variable is not a variable that is random. It is a symbol which is used to quantify the outcomes of a random event. A random variable is assigned a certain value according to the outcomes of the random event.

E.g. tossing a coin is a random event with 2 outcomes. We can take a variable X and assign it values according to the outcomes:

Suppose X = 1 if outcome = heads, X = 2 if outcome = tails

Then X is a random variable. It is a function of the outcome of a random event.

Q.2.

A true random number is one that cannot be determined beforehand. A pseudorandom number is a number generated by software, which "seem" like random numbers but can actually be reproduced, if need be. They are not truly random but they do show some degree of randomness. It takes exponential time to get back the seed value from a pseudorandom sequence. Hence they are good enough to simulate statistically random processes.
